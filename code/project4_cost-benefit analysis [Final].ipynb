```python
# Importing the pre-requisite libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pandas import DataFrame as df
```

```python
# Ensuring our notebook remains tidy
import warnings
warnings.filterwarnings('ignore')
```

```python
# Setting our seaborn style and palette
sns.set_style('darkgrid')
sns.set_palette('icefire')
```

```python
# Importing our predictions (probabilities)
test = pd.read_csv('../assets/test_final.csv')
print(test.shape)
test.head()
```

```python
# Importing our final predictions
pred = pd.read_csv('../kaggle/submission.csv')
print(pred.shape)
pred.head()
```

```python
# Merging probabilities and final predictions
test_pred = test.merge(pred, left_on = 'id', right_on = 'Id')
```

```python
# Visualizing our predicted WNV probabilities
mapdata = np.loadtxt("../assets/map/mapdata_copyright_openstreetmap_contributors.txt")

aspect = mapdata.shape[0] * 1.0 / mapdata.shape[1]
lon_lat_box = (-88, -87.5, 41.6, 42.05)

plt.figure(figsize=(10,16))
plt.imshow(mapdata, cmap=plt.get_cmap('gray'), extent=lon_lat_box, aspect=aspect)

# KDE plot was used for predicted cases
sns.kdeplot(data=test_pred, x='longitude', y='latitude', fill=True, cmap='icefire', alpha = 0.7)

plt.scatter(data=test_pred[test_pred['WnvPresent']!=0], x='longitude', y='latitude', marker='x', 
            color='black', label='WNV Probability')

plt.title('Locations with Risk of WNV', fontsize=18)
plt.legend()
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.tight_layout()
```

## Cost calculations

**Summary of cost items**: Pursuing comprehensive spray efforts across Chicago

* Area of Chicago: 149,800 acres
* Spray used by Chicago: Zenivex E4
* Cost of spray: USD 0.67 per acre, before rebates; supplied by Wellmark International

**Cost of spraying in Chicago**: 149,000 acres $\times$ USD 0.67 per acre = USD 100,366

**Medical Costs in Chicago**

In 2014, data was published in *ScienceDaily* for the West Niles Virus (WNV). These are listed below:

Acute symptoms:
* Initial medical cost:
USD 25,000

* Long term medical cost:
USD 22,000


Fever related symptoms:
* Initial medical cost:
USD 7,500

**Population Affected by WNV in Chicago, 2014**

According to the source above, there were 15 cases of WNV in Chicago. There were 2 serious cases.

* Acute cases: 45,000 x 2 = USD 90,000
* Typical cases: 7,500 x 13 = USD 97,500

**Total estimated cost of medical care**: USD 90,000 + USD 97,500 = USD 187,500

**Productivity costs in Chicago**

Refer to *Table 4*:

* For patients $<$ 60 years of age: USD 191 per day
* For 100% of cases, these equated to 50 work-days missed

**Total cost of lost productivity**: USD 191 $\times$ 50 $\times$ 15 = USD 143,250

**Total cost involved if there are 15 cases with 2 acute cases**: USD 187,500 + USD 143,250 = USD 330,750

**Cost of spraying in Chicago**: 149,800 acres $\times$ USD 0.67 = USD 100,366

## Overall analysis

* From simulation study done in the sources above and cost benefit analysis performed, it can be seen that the cost of spraying entire city with pesticides to prevent West Nile Virus outbreak is less than 15 person infected with West Niles Virus.
* The estimated cost of lost productivity for approx. 15 cases is higher than the cost of spraying the entire city with pesticides.
* Hospital expenses are higher than the cost of spraying the entire city.

Assuming the study performed by pesticide maker for effectiveness above works in controlling the mosquito population as advertised, it can be concluded that spraying the entire city is a more cost effective approach.


## Conclusion

We used **XGBoost** (and tuned hyperparameters) as our final model to predict mosquitoes infection. Using the ROC AUC score as our key metric, we managed to achieve a score of 0.84-85. This score indicates a high-level of class separability and shows that the probability of making a correct class prediction is relatively high if thresholds are tweaked.

**Our recommendations**

1. Based off data we have around the traps, we can infer the presence of clusters, which have higher WNV rates. These clusters / areas where the traps are located should be targeted specifically
2. We observed some **seasonality effects** in our model (i.e. week and month). This means it may be more efficient to start spraying based on seasonal spikes in July-August before they peak later
3. Identify periods of slower wind-speeds as this would likely result in mosquitoes travelling over a larger area then
4. Control mosquito breeding grounds through surveillence and education
